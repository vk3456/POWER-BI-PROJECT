# Bar Charts and Histograms

Bar charts are used to visualize the ordinal data and ordinal is mixup of combinational data and numeric data.

Histograms are commonly used to visualize the correlation spread range (distribution) type of data. Histogram gives you information about the periods and the values for that particular period


### Plotting Bar Chart

```python
plt.bar([10,20,30],[10,20,10],label="Series One",color="red")
plt.bar([1,2,3],[10,20,10],label="Series Two",color="green")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Graph Title")
plt.legend()
```

```python
plt.bar([10,20,30],[10,20,10],label="Series One",color="red")
plt.bar([1,2,3],[10,20,10],label="Series Two",color="green")
plt.plot([1,2,3],[10,20,10],label="LinePlot",color="blue")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Graph Title")
plt.legend()
```

if you want to visualize the distribution then you should go for histogram (specialist)

### Plotting Histogram

```python
from random import randint
voters_age = [randint(18,60) for x in range(50)]
```

```python
ids = list(range(len(voters_age)))
plt.bar(voters_age,ids,color="purple",label="Voters")
plt.xlabel("Age of Voters")
plt.ylabel("Ids of Voters")
plt.title("Voters Age Distribution")
plt.legend()
```

```python
dist = [18,30,35,40,50,60]
plt.hist(voters_age,dist,color="purple", label="Voters", histtype="bar")
plt.xlabel("Age of Voters")
plt.ylabel("Ids of Voters")
plt.title("Voters Age Distribution")
plt.legend()
```

```python
dist = [18,30,35,40,50,60]
plt.hist(voters_age,dist,color="purple", label="Voters", histtype="step")
plt.xlabel("Age of Voters")
plt.ylabel("Ids of Voters")
plt.title("Voters Age Distribution")
plt.legend()
```

# Stack Plot
This type of chart shows the visual information for relational or ordinal type of data

stack graph can not have its own legends


```python
days = [1,2,3,4,5,6,7,8,9]
eating = [2,3,1,5,6,1,2,3,4]
sleeping = [3,1,6,5,7,5,2,3,4]
plt.plot([],[], color="red",label="Earting")
plt.plot([],[], color="green",label="Sleeping")
plt.stackplot(days,eating,sleeping, colors = ["red","green"])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("MY STACK PLOT")
plt.legend()

```

## Stacked Bar Charts
These are generally a good way to represent multiple types of data within a single bar. For example you need to visualize the population of some areas and you want to show the male and female value comparitively.


```python
cities = [
    "New Delhi",
    "Gurugram",
    "Noida",
    "Gaziabaad",
    "Faridabaad"
]
men = [20,35,30,35,27]
women = [25,35,30,34,20]

fig,ax = plt.subplots()
ax.bar(
    cities,
    men,
    0.30,
    label="Men"
)
ax.bar(
    cities,
    women,
    0.30,
    bottom=men,
    label="Women"
)

ax.set_ylabel("Population")
ax.set_title("Population by Gender")
ax.legend()
```

## Grouped Bar Charts

```python
fig,ax = plt.subplots()
x = np.arange(len(cities))
r1 = ax.bar(
    x - 0.30/2,men,0.30,label="Men"
)
r2 = ax.bar(
    x + 0.30/2,women,0.30,label="Women"
)
ax.set_ylabel("Population")
ax.set_title("Population by gender")

ax.set_xticks(x)

ax.set_xticklabels(cities)

ax.legend()

```

```python
apps = ['FACEBOOK','TIKTOK','INSTAGRAM','TWITTER','WHATSAPP']
y_pos = np.arange(len(apps))
downloads = 3+(10 * np.random.rand(len(apps)))
print(apps,y_pos,downloads,sep="\n\n")
```

```python
fig,ax = plt.subplots()

ax.barh(y_pos,downloads,align="center")
ax.set_yticks(y_pos)
ax.set_yticklabels(apps)
ax.invert_yaxis() # top-to-bottom
ax.set_xlabel("Downloads in Millions")
ax.set_title("Trending Apps of Today")
```

# Plotting Categorical Data

```python
activities = ["HANGOUT","GAMING","ME TIME","BIKING","DRIVING"]
male = ["LIKE","LIKE","HATE","LIKE","LIKE"]
female = ["LIKE","HATE","LIKE","LIKE","HATE"]

fig,ax = plt.subplots()
ax.plot(activities,male,label="Male")
ax.plot(activities,female,label="Female")
ax.legend()
```

## Multiple Charts Plotting

```python
fig,axs = plt.subplots(1,3,figsize=(9,3), sharey=True)

axs[0].bar(apps,downloads)
axs[1].scatter(apps,downloads)
axs[2].plot(apps,downloads)
axs[0].set_title("Bar Chart")

axs[2].set_title("Line Chart")
axs[0].tick_params(labelrotation=45)
axs[1].tick_params(labelrotation=45)
axs[2].tick_params(labelrotation=45)
fig.suptitle("Trending Apps")
```

# Seaborn

```python
import seaborn as sns
sns.set(style = "darkgrid") # default theme for all the graphs
```

```python
tips = sns.load_dataset("tips") # loading the data set
```

```python
tips.head()
```

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(
    x = "total_bill",
    y = "tip",
    hue="smoker", # category columns for visualization
    data=tips
)
```

```python
sns.relplot(
    x = "total_bill",
    y = "tip",
    hue = "smoker",
    style = "smoker", # automatically it will pick the diff -2 markers for every unique data point
    data=tips
)
```

```python
sns.relplot(
    x="total_bill",
    y="tip",
    hue="smoker",
    style="time",
    data=tips
)
```

# Visualizing Categorical Data with Seaborn

```python
sns.catplot(x="day", y = "total_bill", data=tips)
```

```python
sns.catplot(x = "time", y="total_bill",data=tips)
```

```python
# who is giving more tip smoker or non smoker
sns.catplot(x="smoker", y="tip", data = tips)
```

```python
# changing the order
sns.catplot(
    x = "smoker",
    y = "tip",
    order = ["No","Yes"],
    data=tips
)
```

```python
# which day is best for dinner and which day is best for lunch

sns.catplot(
    x = "total_bill",
    y = "day",
    hue="smoker",
    data = tips
)
```

```python
sns.catplot(
    x = "total_bill",
    y="day",
    hue="time",
    kind="bar",
    data=tips
)
```

## on the basis of gender visulize the tips over dinner or lunch

```python
sns.catplot(
    x = "sex",
    y = "tip",
    hue="time",
    kind = "bar",
    data = tips
)
```
